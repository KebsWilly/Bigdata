```python
import pandas
data = pandas.read_csv("http://modcom.co.ke/bigdatasept/datasets/Customers.csv")
data
```

```python
print(data.corr())
print(data.describe())

array = data.values
X = array[:, 3:6]  #  
Y = array[:, 6] # 

from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y,test_size = 0.30,
                                                                    random_state=42)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)  # Model is fit with training data, test data is on hideout

predictions = model.predict(X_test)
print(predictions)

from sklearn.metrics import r2_score
print(r2_score(Y_test, predictions))

#amount = model.predict([[50,60,45], [10,30,12],[0,0,0], [20,50,200]])
#print(amount)
```
